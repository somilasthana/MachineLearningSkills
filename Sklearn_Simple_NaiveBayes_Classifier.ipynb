```
"""
Basic Gaussian Naive Bayes
"""
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB
```

```
"""
Yhe assumption is that data from each label is drawn from a simple Gaussian distribution.
"""

X, y = make_blobs(100, 2, centers=2, random_state=42, cluster_std=1.5)
```

```
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
```

```
# if we reduce the std
X, y = make_blobs(100, 2, centers=2, random_state=42, cluster_std=.25)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
```

```
# if we reduce the std
X, y = make_blobs(100, 2, centers=2, random_state=42, cluster_std=6.25)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
```

```
model = GaussianNB()
model.fit(X, y)
```

```
#Generate some data points
rng = np.random.RandomState(0)
Xnew = [-20, -14] + [40, 48] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```

```
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```
